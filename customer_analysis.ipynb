```python
#We must choose MEDIAN to fill NULL values over MEAN as, MEDIAN is ROBUST to OUTLIERS. Mean is average, MEDIAN is arrange ascending and the middle most value, MODE is most frequency
#We will use median but not the median of complete dataset, but categorical median for each category.
```

```python
# Checked the medians of all the accesories by categories
df.groupby('Category')['Review Rating'].median()
```

```python
#FILL NULL VALUES with category medians

df['Review Rating'] = df['Review Rating'].fillna(
    df.groupby('Category')['Review Rating'].transform('median')
)


#transform(): 1) returns same size output as original dataframe, 2)allows row-wise filling. 
# Without transform → mismatch error, With transform → perfect alignment 
```

```python
df['Review Rating'].isnull().sum()  #NO NULL VALUES REMAIN
```

```python
# IT is always a good practice to put your COLUMN NAMES in SNAKE CASING. As our columns now have Upper case letters and spacing between them
#SNAKE CASING is ALL lower case with Underscores

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_') #used loer and replace function to replace all to lowercase and underscores
```

```python
#Change a column name to easier column name
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#  df = df.rename(columns = {'':''}) : yaad rakh, key : value (pair)

#Suppose Age data:

#10, 12, 15, 18, 20, 25, 30, 40
#✅ Using pd.cut(x, 4)

#👉 Range is divided equally:
#(10–17.5), (17.5–25), (25–32.5), (32.5–40)

#✔ Equal range
#❌ Unequal number of values

#✅ Using pd.qcut(x, 4)

#👉 Data is divided into equal groups:

#Group 1 → 2 values  
#Group 2 → 2 values  
#Group 3 → 2 values  
#Group 4 → 2 values  

#✔ Equal number of values
#❌ Unequal ranges

```

```python
#Create a column age group using qcut function so the sales team can get insight of the age brackets that we have.
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

#Create new column age_group
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10) #Check whter it is filled correctly or not
```

```python
# It is easier to work with numerical data so creating a new column where we will convert this string data into numerical type

#Create a column named frequency_mapping

frequency_mapping = {
   'Fortnightly': 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90 
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
